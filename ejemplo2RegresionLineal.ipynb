```python
# Importo las librerias
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import  linear_model
```

```python
dfDatos = pd.read_csv('ingresos-salario/ingreso.csv')
dfDatos
```

```python
# Visualizar y validar si hay regresion lineal
plt.ylabel("Ingresos ($)")
plt.xlabel("Promedio de horas semanales trabajadas (tiempo)")
plt.scatter(dfDatos['horas'], dfDatos['ingreso'], color="blue")
plt.show()
```

```python
# Creador de modelos
regresion = linear_model.LinearRegression()
# Preparar datos en un vector de un solo elemento
horas = dfDatos['horas'].values.reshape((-1, 1))
# Entrenamos el modelo
modelo = regresion.fit(horas, dfDatos['ingreso'])
```

```python
print(f"Interseccion (b) {modelo.intercept_}")
print(f"Pendiente (m) {modelo.coef_}")
```

```python
# Pruebas
entrada = [[39], [40], [43], [44]]
modelo.predict(entrada)
```

```python
# Visualizar y validar si hay regresion lineal
plt.scatter(entrada, modelo.predict(entrada), color="red")
plt.plot(entrada, modelo.predict(entrada), color="black")
plt.ylabel("Ingresos ($)")
plt.xlabel("Promedio de horas semanales trabajadas (tiempo)")
plt.scatter(dfDatos['horas'], dfDatos['ingreso'], color="blue")
plt.show()
```
