```python
# Importar las librerias
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
import pandas as pd
import plotly.express as px
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Carga los datos en dos dataframe
dataIris = load_iris()
dfXDataIris = pd.DataFrame(dataIris['data'])
dfYDataIris = pd.DataFrame(dataIris['target'])
```

```python
# Visualizar la informaci√≥n descriptiva de los datos
print(dfXDataIris.describe())
print(dfYDataIris.describe())
```

```python
# Visualizar la informaci√≥n
pd.plotting.scatter_matrix(dfXDataIris, c=dfYDataIris, diagonal='kde')
```

```python
# Entrena el modelo con 2, 3, 4, 5 y 6 clusters
modeloKmeans = [ KMeans(n_clusters=i, init='random', n_init=10).fit(dfXDataIris) for i in range(2,6) ]
```

```python
# Crea las etiquetas de cada clusters
etiquetas = [f"cluster_{i}" for i in range(2, 6)]
predicciones = np.array([ k.predict(dfXDataIris) for k in modeloKmeans])

for i in range(2, 6):
    pd.plotting.scatter_matrix(dfXDataIris, c=predicciones[i-2])
    plt.suptitle(f"Numero clusters = {i}")
```
